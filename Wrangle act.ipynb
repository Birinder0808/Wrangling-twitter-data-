# Gathering Data

```python
# Importing necessary libraries
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer
import requests
import pandas as pd
import numpy as np
import re
```

```python
# Accessing the account through developer tokens
import tweepy

consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

```python
# Rmoving the Time limiter.
api = tweepy.API(auth_handler=auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
# Reading the twitter archive for accessing the tweets Id
df1 = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
# Grabbing all the tweet Ids into single expression
tweet_ids = df1['tweet_id'].values
```

```python
# Writing a loop to write all the values for tweet ids into json.txt file
count = 0
fails_dict = {}
start = timer()

with open('tweet_json.txt', 'w') as outfile:
    
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# Appending all the information extracted from json.text into a Dataframe
tweet_api = pd.DataFrame(columns = ['tweet_id','favorite_count','retweet_count'])
with open ('tweet_json.txt') as file:
    for line in file:
        status = json.loads(line)
        tweet_id = status['id']
        favorite_count = status['favorite_count']
        retweet_count = status['retweet_count']
        tweet_api = tweet_api.append(pd.DataFrame([[tweet_id,favorite_count,retweet_count]],
                                                 columns = ['tweet_id','favorite_count','retweet_count']))
tweet_api = tweet_api.reset_index(drop = True)
tweet_api
        
```

```python
# Saving the dataframe as csv
tweet_api.to_csv('tweet_api.csv')
```

##### DEFINE (E)

After assessing the data without program I've found that some of the names were incorrect. So I will extract all the names from the text section and create a new series for it.

#####  CODE (E)

```python
#Extracting dog_name from the text column
text_dog_name = []
for text in twitter_archive_enhanced.text:
  name_search = re.search('([Tt]his is\s\w+)', text) or re.search('([Mm]eet\s\w+)', text)
  if name_search:
    # split the string and choosing last word
    name = name_search.group().split()[-1]
    text_dog_name.append(name)
    
text_dog_name
```

```python
#adding text_dog_name column in master_clean data frame
twitter_archive_enhanced['text_dog_name'] = pd.Series(text_dog_name)
```

#####  DEFINE (F)

As we know that the doggo, floofer, pupper and puppo are the four stages of the dogs. But somehow they are all differently categorised in the different columns so we need to extract them all into single column named stage.

#####  CODE (F)

```python
# Replacing the Null Values with space for proper concatanation
twitter_archive_enhanced['doggo'] = twitter_archive_enhanced['doggo'].replace('None','',regex = False)
twitter_archive_enhanced['floofer'] = twitter_archive_enhanced['floofer'].replace('None','',regex = False)
twitter_archive_enhanced['pupper'] = twitter_archive_enhanced['pupper'].replace('None','',regex = False)
twitter_archive_enhanced['puppo'] = twitter_archive_enhanced['puppo'].replace('None','',regex = False)
```

```python
# Adding all the values together
twitter_archive_enhanced['stage'] = twitter_archive_enhanced['doggo']  + twitter_archive_enhanced['floofer']  + twitter_archive_enhanced['pupper'] + twitter_archive_enhanced['puppo']
```

```python
#Handling multiple stages in dog_stage column
twitter_archive_enhanced.loc[twitter_archive_enhanced.stage=='doggopupper', 'dog_stage'] = 'doggo , pupper'
twitter_archive_enhanced.loc[twitter_archive_enhanced.stage=='doggopuppo', 'dog_stage'] = 'doggo , puppo'
twitter_archive_enhanced.loc[twitter_archive_enhanced.stage=='doggofloofer', 'dog_stage'] = 'doggo , floofer'
```

```python
twitter_archive_enhanced.loc[twitter_archive_enhanced.stage=='', 'stage'] = np.nan
```

```python
# Dropping all the Unnecessary Columns
twitter_archive_enhanced.drop(['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_id','retweeted_status_user_id','retweeted_status_timestamp','expanded_urls','doggo','floofer','pupper','puppo','rating_numerator','rating_denominator','name'],axis = 1 , inplace = True)
```
